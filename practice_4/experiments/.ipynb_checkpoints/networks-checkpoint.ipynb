# Networks

## Baseline

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as ss
import ipywidgets as widgets
```

```python
df = pd.read_csv('demand-productivity.csv', skiprows=6).drop(columns=[])
df = df.rename(columns={'[run number]': 'run_num', '[step]': 'step', 'max-demand-users': 'max_demand', 'network': 'net_type', 'mean percent-lockdown': 'perc_lock'})
df.columns
```

```python
dd_values = ['minimal', 'complete', 'random']

dd = widgets.Dropdown(options=dd_values, description='net_type')

ui = widgets.HBox([dd])

def draw_lineplot(net_type):
    filtered = df[df['net_type'] == net_type]
    sns.lineplot(data=filtered, x='step', y='perc_lock', style='max_demand').set(title=f'net_type = {net_type}')
    plt.show()

out = widgets.interactive_output(draw_lineplot, {'net_type': dd})

display(ui, out)
```

Distribution of user demand is double peaked. Threshold when more then 10% users will face power shutdown for different network types:
- random: max-users-demand = 0.6
- complete: max-users-demand = 0.5
- minimal: max-users-demand = 0.3

Conclusions:
1. If user can consume from every distribution station, users will be satisfied for higher demand rate.
2. If we select stations randomly, there is still higher possible demand rate, than for minimal network, where user is connected to only 1 nearest station.

## Own network

```python
df = pd.read_csv('own-dem-prod.csv', skiprows=6).drop(columns=[])
df = df.rename(columns={'[run number]': 'run_num', '[step]': 'step', 'max-demand-users': 'max_demand', 'mean percent-lockdown': 'perc_lock'})
df.columns
```

```python
sns.lineplot(data=df, x='step', y='perc_lock', style='max_demand')
plt.show()
```

Selected parameters:
- users = 150
- generation-stations = 1
- distribution-stations = 5
- demand = double-peaked

To get percent of lockdown near threshold (10 %), choose max demand = 0.3.

```python
df = pd.read_csv('own-n-dist.csv', skiprows=6).drop(columns=[])
df = df.rename(columns={'[run number]': 'run_num', '[step]': 'step', 'destr-n-dist': 'destr_num', 'mean [current-power] of power-stations': 'pow_stat', 'mean percent-lockdown': 'perc_lock'})
df.columns
```

```python
dd_values = ['perc_lock', 'pow_stat']

dd = widgets.Dropdown(options=dd_values, description='net_type')

ui = widgets.HBox([dd])

def draw_lineplot(metric):
    sns.lineplot(data=df, x='step', y=metric, style='destr_num')
    plt.show()

out = widgets.interactive_output(draw_lineplot, {'metric': dd})

display(ui, out)
```

Dependency on number of destroyed stribution stations:
- 0: the percent of locked users just below threshold of 0.1
- 1 - 4: the percent grows quit slowly
- 5: on last station the percent jumps almost twice

Conclusions:
1. Model is very stable because percent of lockdown increases rapidly only when we destroy last station.

```python
base_df = pd.read_csv('base-n-dist.csv', skiprows=6).drop(columns=[])
base_df = base_df.rename(columns={'[run number]': 'run_num', '[step]': 'step', 'destr-n-dist': 'destr_num', 'mean percent-lockdown': 'perc_lock'})
base_df['model_type'] = 'base'
base_df.columns
```

```python
df = df.drop(columns=['pow_stat'])
df['model_type'] = 'own'
```

```python
compare_df = pd.concat([df, base_df]).reset_index(drop=True)
compare_df.head()
```

```python
for destr_num in sorted(compare_df['destr_num'].unique()):
    sns.lineplot(data=compare_df[compare_df['destr_num'] == destr_num], x='step', y='perc_lock', hue='model_type').set(title=f'destr_num = {destr_num}')
    plt.show()
```

Baseline model for comparison is initial random model with max-demand = 0.6 (for 5 stations - near the 10% threshold of lockdown).

Dependance on destroyed distribution stations:
- 0: almost no differences. Implemented model is a bit more stable.
- 1-2: own model is weaker. I think it is because users, already not connected to stations, can consume multiplied power through connected users. And rest power is not enough for connected users. Therefore, less users are satisfied.
- 3-4: users connections becomes important. In imlemented model users can consume through connected users. Therefore, more users are satisfied.
- 5: no connections left. Models give the same result.

To summarise:
1. Implemented own model is more resilient than intial random.
2. It may be weaker, only if we destroy 1 or 2 stations. Because distant users consume lots of energy. Therefore, there isn't enough energy for immediate users of station.

```python
links_df = pd.read_csv('own-n-links.csv', skiprows=6).drop(columns=[])
links_df = links_df.rename(columns={'[run number]': 'run_num', '[step]': 'step', 'destr-n-links': 'destr_num', 'mean percent-lockdown': 'perc_lock'})
links_df['destr_type'] = 'links'
links_df.columns

dist_df = pd.read_csv('own-n-dist.csv', skiprows=6).drop(columns=['mean [current-power] of power-stations'])
dist_df = dist_df.rename(columns={'[run number]': 'run_num', '[step]': 'step', 'destr-n-dist': 'destr_num', 'mean percent-lockdown': 'perc_lock'})
dist_df = dist_df[(dist_df['destr_num'] != 0) & (dist_df['destr_num'] != 5)]
dist_df['destr_type'] = 'dist'
dist_df.columns

compare_df = pd.concat([links_df, dist_df]).reset_index(drop=True)
compare_df.head()
```

```python
# sns split in groups with interval 15
g = sns.FacetGrid(compare_df, col="destr_type")
g.map(sns.lineplot, 'step', 'perc_lock', 'destr_num')
g.add_legend()
plt.show()
```

```python
sns.lineplot(data=links_df, x='step', y='perc_lock', style='destr_num')
plt.show()
```

```python
sns.lineplot(data=dist_df, x='step', y='perc_lock', style='destr_num')
plt.show()
```

Conclusions:
1. Destroying seperate links is not such dangerous, as destroying whole station. Destroying up to 20% of links is less important than destroying 1 station.
2. If we destroy links, we can't predict exact effect, because we may destroy either link between 2 stations, user and station, or user and user. Each case have different effect.
3. We can destroy even 70% of links, but still have quite good level of users satisfaction.
